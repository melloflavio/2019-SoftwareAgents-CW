```python
def create_map_image(map_matrix, tile_dict):
    (n_rows, n_cols) = np.array(map_matrix).shape
    print(n_rows)
    print(n_cols)
    image_size = (n_rows * SIDE_LENGTH, n_cols * SIDE_LENGTH)
    img = Image.new('RGB', image_size)
    
    print(tile_dict)
    
    for i in range(n_rows):
        for j in range(n_cols):
            tile_id = map_matrix[i][j]
            tile_img = tile_dict[tile_id]
            img.paste(tile_img, (i*SIDE_LENGTH, j*SIDE_LENGTH))
            
    return img

        
# img = add_frame(img)
# img.save('output.jpg')
```

```python
def add_frame(img):
    FRAME_WIDTH = 4 # in px
    FRAME_COLOR = 'black'
    
    original_size = img.size
    frame_size = tuple([dim + 2*FRAME_WIDTH for dim in original_size])
    print(frame_size)
    framed_img = Image.new('RGB', frame_size, FRAME_COLOR)
    framed_img.paste(img, (FRAME_WIDTH, FRAME_WIDTH))
    
    return framed_img
```

```python
tl = Image.new('RGB', (248,248), 'white')
framed = add_frame(tl)
plain = Image.new('RGB', (256,256), 'white')

maze_tile_dict = {
    0: plain,
    1: framed,
}
```
